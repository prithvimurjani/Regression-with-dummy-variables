```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn
seaborn.set()
```

```python
raw_data = pd.read_csv('1.03.+Dummies.csv')
#Attendance value is yes if attendance is greater than 75%
raw_data
```

```python
raw_data.describe()
```

```python
data = raw_data.copy()
```

```python
data['Attendance'] = data['Attendance'].map({'Yes':1,'No':0})
```

```python
data
```

```python
data.describe()
```

```python
y = data['GPA']
x1 = data[['SAT','Attendance']]
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(data['SAT'],y,c=data['Attendance'],cmap='RdYlGn_r')
yhat_no = 0.6493 + 0.0014 * data['SAT']
yhat_yes = 0.6493 + 0.0014 * data['SAT'] + 0.2226 
fig = plt.plot(data['SAT'],yhat_no,lw = 2,c='#006837')
fig = plt.plot(data['SAT'],yhat_yes,lw = 2,c='#a50026')
plt.xlabel('SAT',fontsize=20)
plt.ylabel('GPA',fontsize=20)
plt.show()
```

```python
# We notice above that both lines have same slope but different intercepts, this shows that students with 
# higher attendance have higher GPA by 0.2226 on an average
```

Plotting the original regression line

```python
plt.scatter(data['SAT'],data['GPA'],c=data['Attendance'],cmap = 'RdYlGn_r')
yhat_no = 0.6439 + 0.0014*data['SAT']
yhat_yes = 0.8665 + 0.0014*data['SAT']
yhat = 0.0017 * data['SAT'] + 0.275
fig = plt.plot(data['SAT'],yhat_no, lw=2, c='#006837', label ='regression line1')
fig = plt.plot(data['SAT'],yhat_yes, lw=2, c='#a50026', label ='regression line2') 
fig = plt.plot(data['SAT'],yhat, lw=3, c='#4C72B0', label ='regression line')
plt.xlabel('SAT', fontsize = 20),
plt.ylabel('GPA', fontsize = 20),
plt.show()
```
